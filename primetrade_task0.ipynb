```python
import pandas as pd
```

```python
mood_data = pd.read_csv("fear_greed_index.csv")
player_data = pd.read_csv("historical_data.csv")
```

```python
print(mood_data.head())
print(player_data.head())
```

```python
player_data['Timestamp IST'] = pd.to_datetime(player_data['Timestamp IST'], format='%d-%m-%Y %H:%M')
player_data['date'] = player_data['Timestamp IST'].dt.strftime('%Y-%m-%d')
print(player_data[['Timestamp IST', 'date']].head())
```

```python
combined_data = pd.merge(player_data, mood_data, on='date')
print("Combined data")
print(combined_data[['Account', 'date', 'Closed PnL', 'classification']].head())
```

```python
mood_counts = combined_data['classification'].value_counts()
print("Number of Trades in Each Mood")
print(mood_counts)
```

```python
finished_games = combined_data[combined_data['Closed PnL'] != 0.0]
average_money_won = finished_games.groupby('classification')['Closed PnL'].mean()
print(average_money_won)
```

```python
finished_games = finished_games.copy()
finished_games['is_winner'] = finished_games['Closed PnL'] > 0

win_rate = finished_games.groupby('classification')['is_winner'].mean() * 100

print("Win Rate (Percentage of Games Won)")
print(win_rate)
```
